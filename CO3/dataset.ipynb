```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('/content/sample_data/data.csv')
df
```

```python
df['diagnosis'].value_counts()
```

```python
plt.figure(figsize=[17,9])
sb.countplot(df['diagnosis'].value_counts())
plt.show()
```

```python
df.isnull().sum()
```

```python
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
x = df.drop('diagnosis',axis=1)
y = df.diagnosis
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
y = lb.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units=9,kernel_initializer='he_uniform',activation='relu',input_dim=30))
classifier.add(Dense(units=9,kernel_initializer='he_uniform',activation='relu'))
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
classifier.summary()
```

Compiling ANN

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model = classifier.fit(xtrain,ytrain,batch_size=100,epochs=100)
```

```python
y_pred = classifier.predict(xtest)
y_pred = (y_pred>0.5)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(ytest,y_pred)
score = accuracy_score(ytest,y_pred)
print(cm)
print('score is:',score)
```

```python
plt.figure(figsize=[14,7])
sb.heatmap(cm,annot=True)
plt.show()
```

```python
print(model.history.keys())
plt.plot(model.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(model.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

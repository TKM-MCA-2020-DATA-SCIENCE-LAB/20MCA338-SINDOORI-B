```python
webpage
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
name=soup.find_all('div',class_='_4rR01T')
```

```python
name
```

```python
mob_name=[]
for i in range(0,len(name)):
  mob_name.append(name[i].get_text())
```

```python
mob_name
```

```python
ratings=soup.find_all("div",class_='_3LWZlK')
```

```python
ratings
```

```python
rating=[]
for i in range(0,len(ratings)):
  rating.append(ratings[i].get_text())
```

```python
rating
```

```python
price=soup.find_all("div",class_='_30jeq3 _1_WHN1')
```

```python
price
```

```python
prices=[]
for i in range(0,len(price)):
  prices.append(price[i].get_text())
```

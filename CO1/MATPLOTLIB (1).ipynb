```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'*g--')
plt.subplot(1,2,2)
plt.plot(y,x,'*y--')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.9,0.9])
axes2=fig.add_axes([0.15,0.3,0.3,0.3])
axes1.plot(x,y,'g')
axes2.plot(y,x,'g')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Sindhur!!! ')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Christy!!! ')



```

```python
fig,axes=plt.subplots()
axes.plot(x,y,'r--')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('santoor mummy!!!')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(33,10))
for ax in axes:
  ax.plot(x,y,'*g--')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('Sindhur!!!')

```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend(loc=1)
```

```python
f=plt.figure()
axes1=f.add_axes([0.5,0.5,0.9,0.9])
axes1.set_xlabel('X1 label')
axes1.set_ylabel('Y1 label')
axes1.set_title('Graph Main')
axes1.plot(x,y,"g*-")

axes2=f.add_axes([0.6,1,0.2,0.3])   
axes2.set_xlabel('X2 label')
axes2.set_ylabel('Y2 label')
axes2.set_title('Graph1')
axes2.plot(x,y,"r*-")

axes3=f.add_axes([0.9,1,0.2,0.3])   
axes3.set_xlabel('X3 label')
axes3.set_ylabel('Y3 label')
axes3.set_title('Graph2')
axes3.plot(x,y,"y*-")
```

```python
f,axes=plt.subplots()

axes.plot(x,y,'g*-')
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Graph")
```

```python
fig,axes=plt.subplots(1,2)

for ax in axes:
  ax.plot(x,y,'g*-')
  ax.set_xlabel("X label")
  ax.set_ylabel("Y label")
  ax.set_title("Graph")
  fig.tight_layout()
```

```python
fig,axes=plt.subplots(1,2,figsize=(4,6),dpi=100)  
fig.tight_layout()
```

```python
import seaborn as s
```

```python
dataset=s.load_dataset('tips')
dataset.head()
```

```python
s.barplot(x='sex',y='total_bill',data=dataset)
```

```python
import numpy as n
```

```python
s.barplot(x='sex',y='total_bill',data=dataset,estimator=n.std)
```

```python
s.countplot(x='sex',data=dataset)
```

```python
s.boxplot(x='day',y='total_bill',data=dataset,palette='rainbow')
```

```python
s.boxplot(data=dataset,palette='coolwarm',orient='h')
```

```python
s.violinplot(x='day',y='total_bill',data=dataset,palette='rainbow',hue='sex')
```

```python
s.violinplot(x='day',y='total_bill',data=dataset,palette='rainbow',hue='sex',split=True)
```

```python
s.stripplot(x='day',y='total_bill',data=dataset)
```

```python
s.stripplot(x='day',y='total_bill',data=dataset,jitter=True)
```

```python
s.stripplot(x='day',y='total_bill',data=dataset,jitter=True,hue='sex',palette='coolwarm')
```

```python
s.factorplot(x='sex',y='total_bill',data=dataset,kind='bar')
```
